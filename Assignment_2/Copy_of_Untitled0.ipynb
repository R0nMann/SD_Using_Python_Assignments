```python
#3. WAP in python to check whether the year is leap year or not.
year = int(input("Enter a year: "))

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print(year, " is a leap year")
else:
    print(year, " is not a leap year")
```

```python
#4. WAP in python to check whether a triangle is acute, obtuse or right angle triangle only two angles are given by the user.
angle1 = float(input("Enter the 1st angle: "))
angle2 = float(input("Enter the 2nd angle: "))
angle3 = 180 - angle1 - angle2

if angle1 <= 0 or angle2 <= 0 or angle3 <= 0:
    print("Invalid input")
elif angle1 > 90 or angle2 > 90 or angle3 > 90:
    print("The triangle is obtuse")
elif angle1 == 90 or angle2 == 90 or angle3 == 90:
    print("The triangle is right-angled")
else:
    print("The triangle is acute")
```
