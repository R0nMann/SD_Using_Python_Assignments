```python
#1. WAP in python to check given number is Armstrong number or not.
num = int(input("Enter number: "))
temp = num
count = 0
while (temp>0):
  count += 1
  temp = temp // 10
temp = num
total = 0
while (temp>0):
  digit = temp%10
  total += digit**count
  temp = temp//10
if(total==num): print("It is an Armstrong number.")
else: print("It is not an Armstrong number.")
```

```python
#2. WAP in python to read the numbers until -1 is encountered. Find the average of positive and negative numbers entered by the user.
pos_sum = 0
pos_count = 0
neg_sum = 0
neg_count = 0
n = int(input("Enter number: "))
while(n!=-1):
  if(n>0):
    pos_sum += n
    pos_count += 1
  elif(n<0):
    neg_sum += n
    neg_count += 1
  n = int(input("Enter number: "))
if(pos_count > 0):
  pos_avg = pos_sum / pos_count
  print(f"Average of positive numbers = {pos_avg}")
print()
if(neg_count > 0):
  neg_avg = neg_sum / neg_count
  print(f"Average of negative numbers = {neg_avg}")
```

```python
#4. WAP in python to binary number to equivalent decimal number.
bin = (input("Enter a binary number."))
dec = 0
pow = len(bin) - 1
i = 0
while i < len(bin):
  digit = int(bin[i])
  dec += digit*(2**pow)
  pow -= 1
  i += 1
print(f"Decimal of {bin} is {dec}.")
```

```python
#5. WAP to read a character until * is encountered. Also count the number of uppercase, lowercase and numbers entered by the users.
upper_count = lower_count = digits = 0
while True:
  ch = input("Enter a character: ")
  if(ch=="*"):
    break
  elif ch.isupper():
    upper_count += 1
  elif ch.islower():
    lower_count +=1
  elif ch.isdigit():
    digits += 1
print(f"Number of uppercase letters: {upper_count}")
print(f"Number of lowercase letters: {lower_count}")
print(f"Number of digits: {digits}")
```
