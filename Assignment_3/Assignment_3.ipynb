```python
#1. WAP in python to calculate the average of first n natural numbers using for loop.
n = int(input("Enter number of natural numbers: "))
s = 0
for i in range(1,n+1):
  s = s+i
  i = i+1
avg = s/n
print("Average of",n,"natural numbers =",avg)
```

```python
#2. WAP in python to print the multiplication table of n, where n is entered by the user using for loop.
n = int(input("Enter n: "))
for i in range(1,11):
  print(n,"*",i,"=",i*n)
  i = i+1
```

```python
#3. WAP in python to print all the numbers from m-n thereby classify them as even or odd using for loop.
m = int(input("Enter m: "))
n = int(input("Enter n: "))
for i in range(m,n+1):
  if(i%2 == 0): print(i,"is even")
  else: print(i,"is odd")
  i = i + 1
```

```python
#4. WAP in python to print prime numbers within a given range.
low = int(input("Enter lower bound of range: "))
high = int(input("Enter upper bound of range: "))
for n in range(low,high+1):
  isPrime = True
  for i in range(2,int(n**0.5)+1):
    if n%i == 0:
      isPrime = False
      break
  if isPrime:
    print(n)
```

```python
#5. WAP in python to calculate the factorial of a number using for loop.
num = int(input("Enter number: "))
fact = 1
for i in range(1,num+1):
  fact = fact*i
print(f"Factorial of {num} is: {fact}")
```

```python
#6. WAP in python to print the sum of first N fibonacci numbers.
n = int(input("Enter the value of N: "))
a,b = 0,1
sum = 0
if n <= 0:
    print("Please enter a positive integer.")
else:
    for i in range(n):
        sum = sum+a
        a,b = b,a + b
    print(f"Sum of first {n} Fibonacci numbers: {sum}")
```

```python
#7. WAP in python to print the sum of first N tribonacci numbers.
n = int(input("Enter the value of N: "))
a,b,c = 0,0,1
sum = 0
if n <= 0:
    print("Please enter a positive integer.")
else:
    for i in range(n):
        sum = sum+a
        a,b,c = b,c,a+b+c
    print(f"Sum of first {n} Tribonacci numbers: {sum}")
```

```python
#8. WAP in python to calculate the sum of series 1/1 + 2^2/2 + 3^2/3 +...+ n^2/n.[print the series also]
n = int(input("Enter n: "))
sum = 0
for i in range(1,n+1):
  print(f"{i}^2/{i}",end=" + ")
  sum += i**2/i
  i = i+1
print()
print(f"Sum = {sum}")
```

```python
#9. WAP in python to calculate the sum of series 1+2+4+8+16… [print the series also]
n = int(input("Enter n: "))
sum,pow = 0,1
for i in range(1,n+1):
  pow = i**2
  print(f"{pow}",end="+")
  sum += pow
print()
print(f"Sum = {sum}")
```

```python
#10. WAP in python to calculate the sum of series 1/1! +2/2! 3/3!... [print the series also]
import math
n= int(input("Enter n: "))
sum = 0
for i in range(1,n+1):
  print(f"{i}/{i}!",end=" + ")
  sum += i/math.factorial(i)
print()
print(f"Sum = {sum}")
```

```python
#11. WAP in python to calculate the sum of series ((x-1)/x) + 1/2((x-1)/x)^2 + 1/3((x-1)/x)^3...[print the series also]
x = int(input("Enter x: "))
n = int(input("Enter number of terms: "))
sum = 0
for i in range(1,n+1):
  print(f"1/{i}(({x}-1)/{x})^{i}",end=" + ")
  sum += 1/i*((x-1)/x)**i
print()
print(f"Sum = {sum}")
```

```python
#12. WAP in python to calculate the sum of series 1/x + 2/x^2 + 3/x^3+…[print the series also]
x = int(input("Enter x: "))
n = int(input("Enter number of terms: "))
sum = 0
for i in range(1,n+1):
  print(f"{i}/{x}^{i}",end=" + ")
  sum += i/(x**i)
print()
print(f"Sum = {sum}")
```

```python
#13. WAP in python to calculate the value of an investment using for loop.Input an initial value of investment and annual interest, and calculate the value of investment over time.
# Simple Interest Investment Calculator
principal = float(input("Enter the initial investment amount: "))
rate      = float(input("Enter the annual interest rate: "))
years     = int(input("Enter the number of years: "))
print()
print(f"{'Year':<10} {'Interest Earned':>20} {'Total Value':>15}")

for year in range(1, years + 1):
    interest = principal * rate * year / 100
    total = principal + interest
    print(f"{year:<10} Rs{interest:>18,.2f} Rs{total:>13,.2f}")
total_interest = principal * rate * years / 100
final_value    = principal + total_interest
```
